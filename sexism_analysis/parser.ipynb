```python
import pandas as pd
```

```python
ausprimaryraw = pd.read_csv("data/HouseFirstPrefsByCandidateByVoteTypeDownload-27966.csv",header=1)
```

```python
ausprimaryraw[["PartyNm","PartyAb"]].drop_duplicates()
```

```python
parties = ["UAPP","ON","ALP","GRN","LP","NP","LNP","LPNP","CLP","L/NP"]

ausprimaryparsed = pd.DataFrame(columns=
                                ["Division","State","Party","Primary Vote","Name","Gender"])
for seat in ausprimaryraw['DivisionNm'].unique():
    seatraw = ausprimaryraw[ausprimaryraw['DivisionNm']==seat]
    votes_in_seat = seatraw["TotalVotes"].sum()
    for index,row in seatraw.iterrows():
        if row["PartyAb"] not in parties:
            continue
        if pd.isna(row["GivenNm"]):
            name = ' '.join(i.capitalize() for i in "ABHIMANYU KUMAR".split())
        else:
            try:
                name = f"{row["GivenNm"].capitalize()} {row["Surname"].capitalize()}"
            except:
                print(row)
                name = row["Surname"]

        entry = pd.Series([
            seat,
            row["StateAb"],
            row["PartyAb"],
            row["TotalVotes"]/votes_in_seat,
           name,
            "MANUAL"
        ],index=["Division","State","Party","Primary Vote","Name","Gender"])
        ausprimaryparsed.loc[ausprimaryparsed.shape[0]] = entry
```

```python
ausprimaryparsed.to_csv("data/parsed_primaries.csv",index=False)
```

```python
aussenateraw['PartyAb'] = aussenateraw['PartyAb'].apply(lambda x: 'L/NP' if x in ["LP","NP","LNP","LPNP","CLP"] else x)
firstparse = aussenateraw.groupby(['PartyAb',"DivisionNm","StateAb"])['TotalVotes'].sum().reset_index()
```

```python
aussenateparsed = pd.DataFrame(columns=["Division","State","Party","Senate Vote"])
for seat in aussenateraw['DivisionNm'].unique():
    seatraw = firstparse[firstparse['DivisionNm']==seat]
    votes_in_seat = seatraw["TotalVotes"].sum()
    for index,row in seatraw.iterrows():
        if row["PartyAb"] not in parties:
            continue
        entry = pd.Series([
            seat,
            row["StateAb"],
            row["PartyAb"],
            row["TotalVotes"]/votes_in_seat,
        ],index=["Division","State","Party","Senate Vote"])
        aussenateparsed.loc[aussenateparsed.shape[0]] = entry
```

```python
aussenateparsed.to_csv("data/parsed_senate.csv",index=False)
```
